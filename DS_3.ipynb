#**Titanic Data set**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Titanic.csv')
df.head()
```

```python
df.tail()
```

```python
df.describe() # Statistics
```

```python
# Filled Null Values of Age  column with median
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Age']
```

```python
df.columns
```

# Univariate Analysis

```python
# Histogram of Age
sns.histplot(df['Age'], kde = True, bins = 30)
plt.title('Histogram of Age')
```

```python
# Histogram of Fare
sns.histplot(df['Fare'], kde = True, bins = 20)
plt.title('Histogram of Fare')
```

```python
# Countplot of Sex
sns.countplot(df['Sex'])
```

# From the above countplot we can see that the number of males is higher than the number of females

```python
from matplotlib import colors
# Survived Pi Plot
df.Survived.value_counts().plot.pie(autopct = '%1.1f%%', colors = ('lightgreen', 'lightcoral'))
plt.title('Percentage of People Survived and Died')
```

```python
# Box Plot Of Age
sns.boxplot(x = 'Age', data = df)
```

# From the above pie plot it is seen that only 38.4 %  people survived and 61.6% people died

#  Bivariate Analysis

```python
# Barplot for who survived more among Sex
sns.barplot(x = 'Sex', y = 'Fare', data = df)
```

# From the barplot it is seen that females paid more fare as compared to male. So this is can be one potential reason that, females survived more than males

```python
sns.barplot(x = 'SibSp', y = 'Fare', data = df)
```

```python
sns.barplot(x = 'Parch', y = 'Fare', data = df)
```

```python
sns.barplot(x = 'Pclass', y = 'Survived', data = df)
```

```python
sns.barplot(x = 'Pclass', y = 'Fare', data = df)
```

# From the above bar plot it can be concluded that since Pclass 1 paid more fare, it survived the most as compared to other two classes

```python
#Feature Engineering
df['Family_size'] = df['SibSp'] + df['Parch'] + 1
df['Family_size'].head()
```

```python
sns.barplot(y = "Survived", x = "Family_size", data = df)
```

```python
sns.barplot(x = 'Survived', y = 'Fare', data = df)
```

# The people who paid more survived more, as they might have got more facilities

```python
df.columns
```

```python
numerical_cols = ['Fare', 'Survived', 'Age', 'SibSp', 'Parch', 'Pclass']
corr = df[numerical_cols].corr()
sns.heatmap(corr, annot = True)
```

# **Final Conclusion**  - Those who paid more fare had a  higher chance of survival. Females paid more fare as compared to males and hence had a higher chance of survival. In Pclass as well, class 1 paid more fair and hence survived more. From the heatmap it can be seen that parent and children (Parch) survived better than Spouse and Siblings (SibSp). There can be two possible reasons for it, one is that they paid high fare, and second that since parents are older, and hence might have been rescued before.
